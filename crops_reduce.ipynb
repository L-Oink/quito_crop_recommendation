```python
#=================================================================
# THIS NOTEBOOK REDUCES THE crop_recommendation. Dataset TO THE DESIRED INFORMATION
#=================================================================

import pandas as pd
```

```python
crops = pd.read_csv("Crop_recommendation.csv")
display(crops.head())
display(crops.columns)
```

```python
#Drop unnecessary columns
crops_reduced = crops.drop(columns=['N', 'P', 'K', 'ph'])

display(crops_reduced.head()

#Data Analysis

print("\nCROPS SHAPE")
display(crops_reduced.shape)

print("\nDATA TYPES")
display(crops_reduced.dtypes)

print("\nMISSING")
display(crops_reduced.isnull().sum())

print("\nDUPLICATES")
display(crops_reduced.duplicated().sum())

crops_reduced.to_csv("crops_reduced.csv", index=False)
```

The result is data with no duplicates or missing values. Irrelevant columns have been dropped as well.
